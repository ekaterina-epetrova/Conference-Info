# Correlation of Neutralizing Antibody Levels with Vaccine Effectiveness

The correlation between neutralizing antibody levels and vaccine effectiveness is strong, especially in the early period following vaccination. High levels of these antibodies are generally associated with a robust ability to prevent infection or to reduce the severity of the disease. However, as time progresses from the point of vaccination, several studies have shown that these antibody levels can decrease. This decrease does not linearly correlate with a drop in vaccine effectiveness due to several reasons:

## Memory B and T Cells

The immune response to vaccination isn't solely dependent on neutralizing antibodies. Memory B cells and T cells also play a critical role in providing protection. Even if antibody levels wane, these memory cells can quickly spring into action upon re-exposure to the pathogen, producing new antibodies and attacking infected cells, respectively.

## Breadth of Response

The immune system's response to a vaccine is broad, targeting multiple parts of the virus. Even if antibodies against one part of the virus decrease, immunity isn't entirely lost as other components of the immune response can still offer protection.

## Threshold for Protection

The exact level of antibodies required to confer protection can vary, and for some diseases, even relatively low levels of antibodies can be protective. Additionally, the correlation between neutralizing antibody titers and protection can be nonlinear

African Swine Fever (ASF) is a viral disease affecting domestic pigs and wild boars, leading to significant economic losses in the swine industry. Understanding the host immune response, particularly the B cell repertoire, during ASFV infection is crucial for developing effective diagnostic and therapeutic strategies. Profiling the B cell repertoire can provide insights into the diversity and dynamics of antibody responses against ASFV, aiding in the identification of vaccine targets.

3. Methodology:
a. Sample Collection: Blood samples will be collected from ASFV-infected pigs at different time points post-infection (e.g., pre-infection, acute phase, convalescent phase).
A very high viral titer can be observed in various lymphatic organs, especially spleen, as demonstrated before.

b. B Cell Isolation: Peripheral blood mononuclear cells (PBMCs) will be isolated using density gradient centrifugation, followed by isolation of B cells using magnetic bead-based sorting.

c. BCR Sequencing: Total RNA will be extracted from sorted B cells, and B cell receptor (BCR) sequences will be amplified using multiplex PCR targeting variable (V) and joining (J) gene segments. High-throughput sequencing will be performed to obtain BCR repertoire data.

d. Bioinformatic Analysis: BCR sequences will be processed to identify clonal expansions, determine CDR3 lengths, and calculate diversity metrics. Phylogenetic analysis will be conducted using metrics such as Levenshtein distance to infer evolutionary relationships among BCRs.

e. Functional Relevance Analysis: Integration of BCR sequencing data with functional assays, such as antibody binding assays or virus neutralization assays, will be performed to assess the functional relevance of identified BCRs and antibodies.

6. Future Directions:
